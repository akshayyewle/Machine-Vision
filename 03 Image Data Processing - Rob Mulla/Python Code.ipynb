```python
import pandas as pd
import numpy as np 
import glob as glob
import matplotlib.pyplot as plt 
import cv2
```

# **Create Globs**

```python
# Create Cat Files Glob
cat_files = glob.glob('Input Data/test_set/test_set/cats/*.jpg')
print('No. Of Cat Files: ',len(cat_files))

# Create Dog Files Glob
dog_files = glob.glob('Input Data/test_set/test_set/dogs/*.jpg')
print('No. Of Dog Files: ',len(dog_files))
```

# **Read Images**

```python
# Read Image File Using Matplotlib 
Image_mpl = plt.imread(cat_files[0])
Image_mpl
```

```python
# Read Image File Using CV2 
Image_cv2 = cv2.imread(cat_files[0])
Image_cv2
```

```python
# Compare Image Size
# (Height,Width,Channel)
print(Image_mpl.shape,Image_cv2.shape)
```

```python
# Get Max Intensity Of Image
Image_mpl.max()
```

```python
# Convert Pixel Intensity To 0-1 Scale
Image_mpl.max() / 255
```

# **Create Histogram**

```python
# Flatten A Array
Image_mpl.flatten()
```

```python
pd.Series(Image_mpl.flatten()).plot(kind='hist',bins=50)
```

# **Display An Image**

```python
fig,ax = plt.subplots()
ax.imshow(Image_mpl)
ax.axis('off')
plt.show()
```

```python
# Access Red Band
plt.imshow(Image_mpl[:,:,0],cmap='Reds')
```

```python
fig,axs = plt.Subplot(1,3)
axs[0].imshow(Image_mpl[:,:,0],cmap='Reds')
axs[1].imshow(Image_mpl[:,:,1],cmap='Greens')
axs[2].imshow(Image_mpl[:,:,2],cmap='Blues')
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
plt.show()
```
