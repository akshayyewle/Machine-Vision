```python
Zip_File = ZipFile('intel-image-classification.zip')

print('No. Of File Found: ',len(Zip_File.namelist()))
print('Files Found: ','\n',Zip_File.namelist())
```

```python
for i in range(len(Zip_File.namelist())):
    if i < 1:
        filepath = Zip_File.namelist()[i]
        print(filepath.split('/')[-2::-1])
```

```python
filepath
```

```python
for filepath in Zip_File.namelist():

    # Create Folder For Each Item ID
    Folder_Path = f"{Folder_Input}/{filepath.split('/')[0]}/{filepath.split('/')[1]}" 
    if not os.path.exists(Folder_Path):
        os.makedirs(Folder_Path,exist_ok=True)
        print('New Item Data Folder Created!')
    else:
        File_Name = filepath.split('/')[-1]
        File_Path = os.path.join(Folder_Path,File_Name)
        file_content = Zip_File.read(filepath)
        with open(File_Path, 'wb') as f:
            f.write(file_content)
            # print(f"{File_Name_PV} Downloaded !")
```
